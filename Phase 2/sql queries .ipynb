### 1) Display all details from both tables where movie names are the same.

```python
select_query ='''select Table2.movie_name,Table2.star1,Table2.star2,Table2.votes,Table2.genre1,Table2.genre2,Table2.genre3,Table2.popularity,Table2.certification,Table2.gross ,Table1.director,Table1.duration,Table1.movie_year,Table1.rating from Table1
inner join Table2 
on Table1.movie_name=Table2.movie_name; '''
df = pd.read_sql_query(select_query, db)
```

```python
df
```

### 2) Display all movie names, Director, ratings, and gross collection where the genre is action.

```python
select_query ='''SELECT Table1.movie_name, Table1.director,Table2.genre1, Table1.rating, Table2.gross
FROM Table1
JOIN Table2 ON Table1.movie_name = Table2.movie_name
WHERE Table2.genre1 = 'Action' '''

df = pd.read_sql_query(select_query, db)
```

```python
df
```

## 3)Display all details from both tables with the highest gross collection. 

```python
select_query='''select Table1.movie_name,Table2.star1,Table2.star2,Table2.votes,Table2.genre1,Table2.popularity,Table2.certification,Table2.gross ,Table1.director,Table1.duration,Table1.movie_year,Table1.rating from Table1
inner join Table2 
on Table1.movie_name=Table2.movie_name
where gross= (select max(gross) from Table2)'''

df=pd.read_sql_query(select_query,db)
```

```python
df
```

### Display all details from both tables with the highest ratings

```python
select_query='''select Table1.movie_name,Table2.star1,Table2.star2,Table2.votes,Table2.genre1,Table2.popularity,Table2.certification,Table2.gross ,Table1.director,Table1.duration,Table1.movie_year,Table1.rating from Table1
inner join Table2 
on Table1.movie_name=Table2.movie_name
where rating= (select max(rating) from Table1)'''
df=pd.read_sql_query(select_query,db)
```

```python
df
```

### Display all details from both tables with the lowest gross collection and lowest ratings

```python
select_query='''select Table1.movie_name,Table2.star1,Table2.star2,Table2.votes,Table2.genre1,Table2.popularity,Table2.certification,Table2.gross ,Table1.director,Table1.duration,Table1.movie_year,Table1.rating from Table1
inner join Table2 
on Table1.movie_name=Table2.movie_name
where rating= (select min(rating) from Table1) or gross=(select min(gross) from Table2)'''
df=pd.read_sql_query(select_query,db)
```
