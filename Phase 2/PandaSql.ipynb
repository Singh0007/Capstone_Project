### Make subquery :



#### Display all details from both tables where movie names are the same.

```python
merged_df = pd.merge(df1, df2, on='movie_name', how='inner')
merged_df
```

#### Display all movie names, Director, ratings, and gross collection where the genre is action.


```python
merged_df = df1.merge(df2, on='movie_name', how='inner')
# Filter rows where genre1 is 'Action'
filtered_df = merged_df[merged_df['genre1'] == 'Action']

# Concatenate the desired columns from df1 and df2
result = pd.concat([filtered_df[['movie_name', 'director', 'rating','genre1']], filtered_df['gross']], axis=1)
result
```

#### Display all details from both tables with the highest gross collection.


```python
merged_df = df1.merge(df2, on='movie_name', how='inner')

# Find the row(s) with the highest gross collection
max_gross = merged_df['gross'].max()

highest_gross_movies = merged_df[merged_df['gross'] == max_gross]
highest_gross_movies
```

#### Display all details from both tables with the highest ratings


```python
merged_df = df1.merge(df2, on='movie_name', how='inner')

# Find the row(s) with the highest gross collection
max_ratings = merged_df['rating'].max()

highest_rating = merged_df[merged_df['rating'] == max_ratings]
highest_rating
```

#### Display all details from both tables with the lowest gross collection and lowest ratings

```python
merged_df = df1.merge(df2, on='movie_name', how='inner')

# Find the row(s) with the lowest gross collection and lowest ratings
min_gross = merged_df['gross'].min()
min_rating = merged_df['rating'].min()

#gros=merged_df[(merged_df['rating'] == min_rating)]
                   
lowest_gross_rating_movies = merged_df[(merged_df['gross'] == min_gross) | (merged_df['rating'] == min_rating)]
lowest_gross_rating_movies
#min_gross
```
