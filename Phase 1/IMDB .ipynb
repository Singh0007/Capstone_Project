```python
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

def scrape_movie_genres(url):
    # Replace 'path_to_web_driver' with the actual path to your web driver executable
    driver = webdriver.Chrome()
    driver.get(url)

    genre1, genre2, genre3 = [], [], []
    wait = WebDriverWait(driver, 10)

    for page in range(34):
        movie_containers = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "lister-item-content")))

        for container in movie_containers:
            # Extracting genres using BeautifulSoup
            data = container.get_attribute('outerHTML')
            soup = BeautifulSoup(data, 'html.parser')
            genre_element = soup.find("span", class_="genre")
            if genre_element:
                genres = genre_element.get_text(strip=True).split(', ')
                # Now 'genres' will contain the list of genres for the current movie

                # Append the genres to the respective lists
                if len(genres) > 0:
                    genre1.append(genres[0])
               #@if len(genres) > 1:
               #@    genre2.append(genres[1])
                
                genre2.append(genres[1] if len(genres) > 1 else np.nan)
                    
                    
                # Append an empty string to genre3 if there is no value
                genre3.append(genres[2] if len(genres) > 2 else np.nan)

        # Navigate to the next page
        next_button = driver.find_element(By.CLASS_NAME, "lister-page-next")
        next_button.click()

    driver.quit()

    return genre1, genre2, genre3

# URL of the first page
url_base = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"

# Scrape movie genres from the first four pages and append to genre1, genre2, and genre3
genre1, genre2, genre3 = scrape_movie_genres(url_base)

# Print the scraped genres
#print("Genre 1:", genre1)
#print("Genre 2:", genre2)
#print("Genre 3:", genre3)
```

```python
def scrape_votes(url, num_pages=34):
    all_votes = []

    for page in range(num_pages):
        # Send an HTTP GET request to the URL of the page
        response = requests.get(url)

        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all the movie containers on the page
        movie_containers = soup.find_all('div', class_='lister-item-content')

        for container in movie_containers:
            # Extracting votes
            vote_element = container.find('p', class_='sort-num_votes-visible')
            if vote_element:
                votes_text = vote_element.find('span', attrs={'name': 'nv'}).text
                # Now 'votes_text' will contain the number of votes for the current movie
            else:
                votes_text = np.nan  # Append an empty string if there are no votes
            all_votes.append(votes_text)

        # Find the URL for the next page
        next_button = soup.find('a', class_='lister-page-next')
        if next_button:
            next_page_url = 'https://www.imdb.com' + next_button['href']
            url = next_page_url
        else:
            # Break the loop if there are no more pages
            break

    return all_votes

# URL of the first page
url_base = "https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&ref_=adv_prv"

# Scrape votes for the next 2 pages
votes = scrape_votes(url_base, num_pages=34)

# Print the scraped votes
#print("Votes:", votes)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_certifications(url, num_pages=34):
    all_certifications = []

    for page in range(num_pages):
        # Send an HTTP GET request to the URL of the page
        response = requests.get(url)

        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all the movie containers on the page
        movie_containers = soup.find_all('div', class_='lister-item-content')

        for container in movie_containers:
            # Extracting certifications
            certification_element = container.find('p', class_='text-muted')
            if certification_element:
                certification_span = certification_element.find('span', class_='certificate')
                if certification_span:
                    certification_text = certification_span.text.strip()
                else:
                    certification_text = np.nan  # Append an empty string if certification details are not available
            else:
                certification_text = np.nan  # Append an empty string if certification details are not available

            all_certifications.append(certification_text)

        # Find the URL for the next page
        next_button = soup.find('a', class_='lister-page-next')
        if next_button:
            next_page_url = 'https://www.imdb.com' + next_button['href']
            url = next_page_url
        else:
            # Break the loop if there are no more pages
            break

    return all_certifications

# URL of the first page
url_base = "https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&ref_=adv_prv"

# Scrape certifications for the next 2 pages
certifications = scrape_certifications(url_base, num_pages=34)

# Print the scraped certifications
#print("Certifications:", certifications)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_gross_collection(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    gross_details = []

    movie_containers = soup.find_all("div", class_="lister-item-content")

    for container in movie_containers:
        gross_element = container.find("span", attrs={"name": "nv"})

        if gross_element:
            gross_collection = gross_element["data-value"]
            movie_title = container.find("h3", class_="lister-item-header").find("a").text
            gross_details.append((movie_title, gross_collection))

    return gross_details

base_url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1&page={}"
total_pages = 34

all_gross_details = []

for page_number in range(1, total_pages + 1):
    url = base_url.format(page_number)
    gross_details = scrape_gross_collection(url)
    all_gross_details.extend(gross_details)

# Print the scraped gross collection details from 2 pages
for movie_title, gross_collection in all_gross_details:
    gross.append(f"${gross_collection}")
    #print(f"${gross_collection}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the content with BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find all the movie containers on the page
movie_containers = soup.find_all("div", class_="lister-item-content")

# Lists to store certificate ratings
certificates = []

# Loop through each movie container and extract certificate ratings
for container in movie_containers:
    certificate_element = container.find("span", class_="certificate")

    if certificate_element:
        certificate = certificate_element.text.strip()
        # Replace '133 min' with 'Not Rated'
        if certificate == '133 min':
            certificate = 'Not Rated'
        certificates.append(certificate)

# Ensure the length of the dataset is 50
while len(certificates) < 50:
    certificates.append('Not Rated')

# Take the first 50 values if the dataset has more than 50 entries
certificates = certificates[:50]

# Create a DataFrame to store the scraped data
#df = pd.DataFrame(certificates, columns=["certificate"])

#print(df)
```
